## 1.1交换

## 1.2回文

回文的定义：”回文” 就是正读倒读都一样的。

如奇数个：”98789”，这个数字正读是”98789” 倒读也是”98789”。

偶数个数字”3223”也是回文数。

字母 “abcba” 也是回文。

判断一个字符串是否是回文字符串，是打印True， 不是打印False

```python
def is_palindrome(s):
    return s == s[::-1]
s = "abcba"
print(is_palindrome(s))
s = 12321
print(is_palindrome(str(s)))
s = 123321
print(is_palindrome(str(s)))
s = "asdfasdf"
print(is_palindrome(s))
```

## 1.4拼接字符串

有个列表 [“hello”, “world”, “yoyo”]如何把把列表里面的字符串联起来，

得到字符串 “hello_world_yoyo”

```python
str_list = ['hello', 'world', 'yoyo']
s = '_'.join(str_list)
print(s)
```

## 1.6九九乘法表

打印99乘法表

```python
for i in range(1, 10):
    for j in range(1, i + 1):
        print(f"{j}*{i}={i*j}", end="\t")
    print()
```

## 1.7字符下标

找出单词 “welcome” 在 字符串”Hello, welcome to my world.” 中出现的位置，找不到返回-1

从下标0开始索引

```python
s = "Hello, welcome to my world."
a = s.find("welcome")
print(a)
s = "Hello, hello, hello world"
a = s.find("welcome")
print(a)
```

### python中的find方法简单使用

基本语法：

str.find(sub[, start[, end]])

参数说明：

* sub：要查找的子字符串
* start（可选）：开始查找的位置，默认为 0
* end（可选）：结束查找的位置，默认为字符串末尾

返回值：

* 如果找到子字符串，返回第一次出现的位置索引（从 0 开始）
* 如果找不到，返回 -1

注意事项：

* find() 只返回第一次出现的位置
* 如果需要查找所有出现位置，需要循环调用或使用其他方法
* 类似方法：index() 功能相同，但找不到时会抛出 ValueError 异常，而不是返回 -1

```python
s = "Hello, welcome to my world."
print(s.find("welcome"))  # 输出: 7，下标从0开始
print(s.find("Python"))   # 输出: -1，没找到返回-1

# 指定起始位置
s = "hello hello hello"
print(s.find("hello"))      # 输出: 0（第一个）
print(s.find("hello", 1))   # 输出: 6（从索引1开始找）
print(s.find("hello", 7))   # 输出: 12（从索引7开始找）
```

## 1.8 统计字符出现的次数

统计字符串“Hello, welcome to my world.” 

中字母w出现的次数

统计单词 my 出现的次数

```python
import re
from collections import Counter

s = "Hello, welcome to my world."
c = Counter(s)
print(c['w'])
s = re.split(r"[\s,]", s)
print(s)
c = Counter(s)
print(c['my'])
```
