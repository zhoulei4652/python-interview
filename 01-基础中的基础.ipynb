### python中的find方法简单使用

基本语法：

str.find(sub[, start[, end]])

参数说明：

* sub：要查找的子字符串
* start（可选）：开始查找的位置，默认为 0
* end（可选）：结束查找的位置，默认为字符串末尾

返回值：

* 如果找到子字符串，返回第一次出现的位置索引（从 0 开始）
* 如果找不到，返回 -1

注意事项：

* find() 只返回第一次出现的位置
* 如果需要查找所有出现位置，需要循环调用或使用其他方法
* 类似方法：index() 功能相同，但找不到时会抛出 ValueError 异常，而不是返回 -1

````python
s = "Hello, welcome to my world."
print(s.find("welcome"))  # 输出: 7，下标从0开始
print(s.find("Python"))   # 输出: -1，没找到返回-1

# 指定起始位置
s = "hello hello hello"
print(s.find("hello"))      # 输出: 0（第一个）
print(s.find("hello", 1))   # 输出: 6（从索引1开始找）
print(s.find("hello", 7))   # 输出: 12（从索引7开始找）
````

## 2.1水仙花数
如果一个 3 位数等于其各位数字的立方和，则称这个数为水仙花数。

例如：153 = 1^3 + 5^3 + 3^3，因此 153 就是一个水仙花数

那么问题来了，求1000以内的水仙花数（3位数）

````python
import math
for i in range(100, 1000):
    a, b, c = str(i)
    a = int(a)
    b = int(b)
    c = int(c)
    if math.pow(a, 3) + math.pow(b, 3) + math.pow(c, 3) == i:
        print(i)
    
````

## 2.2完全数
如果一个正整数等于除它本身之外其他所有除数之和，就称之为完全数。

例如：6是完全数，因为6 = 1+2+3；

下一个完全数是28 = 14+7+4+2+1。

求1000以下的完全数

````python
def get_chushu_list(n):
    chushu_list = []
    for i in range(1, n):
        if n % i == 0:
            chushu_list.append(i)
    return chushu_list

for i in range(1, 10000):
    chushu_list = get_chushu_list(i)
    # print(i, chushu_list)
    if sum(chushu_list) == i:
        print(i)
````

定义一个函数：计算 1-n 之间的所有 5 的倍数之和，默认计算 1-100 （ n 是 一个整数）

````python
def sum_5(n):
    result = []
    m = 1
    while m * 5 <= n:
        result.append(m * 5)
        m += 1
    return result
sum(sum_5(100))
````

## 2.11斐波那契数列1、1、2、3、5、8、13 …..
已知一个数列：1、1、2、3、5、8、13、...的规律为从 3 开始的每一项都等于其前两项的和，这是斐波那契数列。
求满足规律的 100 以内的所有数据

````python
def fibonacci(n):
    fib_seq = [1, 1]
    while True:
        next_fib = fib_seq[-1] + fib_seq[-2]
        if next_fib > n:
            break
        fib_seq.append(next_fib)
    return fib_seq
fibonacci(1000)
````

### sorted方法用法的简单介绍

sorted() 是 Python 内建的通用排序函数，用来返回一个新的已排序序列，不会修改原对象。常见参数：

* iterable：可迭代对象，如列表、元组、字符串、集合、字典（默认按键排序）。
* key=None：可选的函数，用于指定排序的比较键，比如 key=len、key=str.lower。
* reverse=False：是否按降序排列，设为 True 时翻转结果。

````python
# 基本排序
sorted([3, 1, 4])                # [1, 3, 4]

# 自定义 key
words = ["apple", "Banana", "cherry"]
sorted(words, key=str.lower)     # ['apple', 'Banana', 'cherry']

# 按长度排序
sorted(words, key=len)           # ['apple', 'Banana', 'cherry']

# 降序
sorted([3, 1, 4], reverse=True)  # [4, 3, 1]

# 排序字典的 key
sorted({"b": 2, "a": 1})         # ['a', 'b']

# 使用自定义函数或者lambda表达式
def by_length(item):
    return len(item)

words = ["python", "AI", "interview"]
sorted(words, key=by_length)     # 结果: ['AI', 'python', 'interview']
````

### sorted方法是怎么调用的key指定的函数的？

sorted() 在内部会对可迭代对象的每个元素调用一次 key 函数，把返回值作为排序依据，而不是直接比较元素本身。

执行流程：

- 遍历所有元素，对每个元素调用 key(element)
- 用这些返回值构建一个"比较键"列表
- 按比较键排序，同时记住元素原始位置
- 返回按比较键顺序重排后的元素列表

```python
words = ["python", "AI", "interview"]
sorted(words, key=len)
```

实际执行：

- len("python") → 6
- len("AI") → 2
- len("interview") → 9

按 [2, 6, 9] 排序 → 结果 ["AI", "python", "interview"]

### key指定的函数只能有一个参数？

是的，key 函数必须只接收一个参数——当前被排序的元素。

这是因为 sorted() 的内部机制决定的：它遍历每个元素时，只会传递该元素本身给 key 函数。

如果需要多个参数怎么办？可以用这些方法绕过限制：

1. 使用 lambda 捕获外部变量

```python
base_price = 100
products = [{"name": "A", "discount": 0.9}, {"name": "B", "discount": 0.8}]
sorted(products, key=lambda p: p["discount"] * base_price)
```

2. 使用 functools.partial 预填充参数
```python
from functools import partial

def calc_score(weight, item):
    return item["value"] * weight

items = [{"value": 10}, {"value": 5}]
sorted(items, key=partial(calc_score, weight=2))  # weight 固定为 2
```

3. 用类或闭包携带状态
```python
class ScoreKey:
    def __init__(self, multiplier):
        self.multiplier = multiplier
    
    def __call__(self, item):
        return item * self.multiplier

sorted([3, 1, 4], key=ScoreKey(10))  # 返回 [1, 3, 4]
```

本质都是把"额外参数"提前绑定或封装，让 key 函数看起来只需要一个参数。

## 3.10 去重保留顺序

将列表中的重复值取出(仅保留第一个)，要求保留原始列表顺序

如a=[3, 2, 1, 4, 2, 6, 1]    输出[3, 2, 1, 4, 6]

````python
# 方法1: 原始方法 (清晰易懂，但效率较低)
a = [3, 2, 1, 4, 2, 6, 1]
result = []
for item in a:
    if item not in result:  # list的in操作是O(n)
        result.append(item)
print("方法1:", result)

# 方法2: 使用dict.fromkeys() (最简洁高效，Python 3.7+字典保持插入顺序)
a = [3, 2, 1, 4, 2, 6, 1]
result = list(dict.fromkeys(a))
print("方法2:", result)

# 方法3: 使用set追踪已见元素 (高效，适合大列表)
a = [3, 2, 1, 4, 2, 6, 1]
seen = set()
result = []
for item in a:
    if item not in seen:  # set的in操作是O(1)
        seen.add(item)
        result.append(item)
print("方法3:", result)
````

## 3.23列表查找两数之和

给定一个整数数组nums 和一个目标值target ，请你在该数组中找出和为目标值的那两个整数，并返回他们的数组下标。

示例:

给定nums=[2，7，11，15]，target=9

因为nums[0] + nums[1] =2+7 = 9

````python
nums=[2, 0, 11, 15, 7]
target = 9

for index_i in range(len(nums)):
    i = nums[index_i]
    for index_j in range(len(nums[index_i + 1:])):
        j = nums[index_i + 1 + index_j]
        if i + j == target:
            print(index_i, index_i + 1 + index_j)
            break
````

## 5.2 冒泡排序

a = [11, 2, 33, 1, 5, 88, 3]

冒泡排序：依次比较两个相邻的元素，如果顺序（如从小到大、首字母从A到Z）,错误就把他们交换过来

````python
a = [11, 2, 33, 1, 5, 88, 3]

# 冒泡排序实现
def bubble_sort(arr):
    n = len(arr)
    # 外层循环控制排序轮数
    for i in range(n):
        # 内层循环进行相邻元素比较和交换
        # 每轮排序后，最大的元素会"冒泡"到末尾
        # n-i-1 是因为每轮排序后最后i个元素已经有序
        for j in range(0, n - i - 1):
            # 如果前一个元素大于后一个元素，交换它们
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr

# 调用冒泡排序
sorted_a = bubble_sort(a.copy())  # 使用copy()避免修改原列表
print("原始列表:", a)
print("排序后:", sorted_a)
````

## 5.19 如何判断一个字符串有没有重复字符

判断一个字符串是否包含重复字符。例如：“hello”就包含重复字符‘l’，而“world”就不包含重复字符, 有重复打印True, 没重复打印False

````python
from collections import Counter

a = "hello"
b = "world"

def is_contains_duplicate(s):
    counter = Counter(s)
    result = list(filter(lambda x: counter[x] > 1, counter))
    return len(result) > 0

print(is_contains_duplicate(a))
print(is_contains_duplicate(b))
````

## 5.20 找出一个字符串中子串不含有重复字符的最长子串（子串）

```
给定一个字符串，请你找出其中不含有重复字符的最长子串的长度。
示例1:
输入:"abcabcbb"  输出: 3
解释:因为无重复字符的最长子串是"abc"， 所以其长度为3。
示例2:
输入: "bbbbb"   输出: 1
解释:因为无重复字符的最长子串是”b”， 所以其长度为1。
示例3:
输入: "pwwkew"  输出: 3
解释:因为无重复字符的最长子串是"wke"， 所以其长度为3。
请注意，你的答案必须是子串的长度，"pwke"是一个子序列，不是子串。
```

````python
s1 = "abcabcbb"
s2 = "bbbbb"
s3 = "pwwkew"

def max_duplicate_length(s):
    temp_s = ""
    duplicate_length = 0
    max_length = 0
    for character in s:
        if temp_s == "":
            temp_s = temp_s + character
        else:
            if character in temp_s:
                duplicate_length = len(temp_s)
                temp_s = character
                if duplicate_length > max_length:
                    max_length = duplicate_length
            else:
                temp_s = temp_s + character
    return max_length

print(max_duplicate_length(s1))
print(max_duplicate_length(s2))
print(max_duplicate_length(s3))
            
````

## 5.21 一个字符串中所有子串是回文的次数(子串)

回文是指正序（从左向右）和倒序（从右向左）读都是一样的。

例如：121 ，abcdedcba，123321等都是回文

这种的字符串“ABCABADCSABBAUYIIYU”找出回文出现的次数

子串回文是：’BB’, ‘II’,’ABA’,’ABBA’, ‘YIIY’,’UYIIYU’  总共有6个

````python
s = "ABCABADCSABBAUYIIYU"
count = 0
temp = ""
for i in range(len(s)):
    character = s[i]
    other = s[i+1:]
    other_len = len(other)

    start = (other_len+1) * -1
    end  = -1
    
    for other_len in range(start, end):
        if end == -1:
            temp = other[start:]
        else:
            temp = other[start:end+1]
        temp = character + temp
        print(temp)
        if len(temp)>1 and temp == temp[::-1]:
            count = count + 1
        end = end - 1
print(count)
````

## 5.22 找出一个列表中，所有出现的连续数（栈）

连续的数字是指：123， 456， 78 这种，可以是连续的2个，也可以是多个，135 这种是不连续的。

于是可以知道连续的数字是[1,2,3]，[5,6,7,8], [98,99,100,101]

````python
a=[1,2,3,8,6,7,5,10,16,98,99,100,101,77,9]
max_num = max(a)
min_num = min(a)
print(max_num, min_num)
result = []
temp = []
for item in range(min_num, max_num+1):
    
    if item in a:
        temp.append(item)
    else:
        result.append(temp)
        temp = []
if len(temp) > 0:
    result.append(temp)

result = list(filter(lambda x: len(x)>1, result))
print(result)
````
