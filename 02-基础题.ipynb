### 2 a=1,b=2,不用中间变量交换 a 和b 的值？

````python
a, b = 1, 2
print(a, b)
a, b = b, a
print(a, b)
````

### 3 print 调用Python 中底层的什么方法?

print(obj),print方法会先调用obj的__str__()方法，如果没有__str__()方法那么会调用__repr__()方法，然后调用sys.stdout.write()方法，这个方法用于向标准输出到控制台。
